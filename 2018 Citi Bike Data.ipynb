```python
import pandas as pd
```

```python
#create dataframes for 2018 Q1-Q4 data daily ridership and membership data

q1_2018_df = pd.read_csv("Data/2018Q1_Ridership_Membership_Bike_Data.csv")
q2_2018_df = pd.read_csv("Data/2018Q2_Ridership_Membership_Bike_Data.csv")
q3_2018_df = pd.read_csv("Data/2018Q3_Ridership_Membership_Bike_Data.csv")
q4_2018_df = pd.read_csv("Data/2018Q4_Ridership_Membership_Bike_Data.csv")
```

```python
#view sample data
q3_2018_df.head()
```

```python
#check data types
q2_2018_df.dtypes
```

```python
#create a list of df to use concat to combine data
frames2018 = [q1_2018_df,q2_2018_df,q3_2018_df,q4_2018_df]
```

```python
#perform pd.concat and preview files
ridership_membership_2018 = pd.concat(frames2018)
ridership_membership_2018.head()
```

```python
#preview last 20 rows to check data
ridership_membership_2018.tail(10)
```

```python
#create a new column for Change in Annual Members, that takes the difference between total annual members from previous row
ridership_membership_2018["Change in Annual Members"] = ridership_membership_2018["Total Annual Members (All Time)"].diff()
ridership_membership_2018.head()
```

```python
ridership_membership_2018.dtypes
```

```python
#send our new combined dataframe to a new csv file
ridership_membership_2018.to_csv("Data/ridership_membership_2018.csv", index=False)
```
