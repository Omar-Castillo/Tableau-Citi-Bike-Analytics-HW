```python
import pandas as pd
```

```python
#create a df for each month of JC 2018 trip data
jan_trip_df = pd.read_csv("Data/JC-201801-citibike-tripdata.csv")
feb_trip_df = pd.read_csv("Data/JC-201802-citibike-tripdata.csv")
mar_trip_df = pd.read_csv("Data/JC-201803-citibike-tripdata.csv")
apr_trip_df = pd.read_csv("Data/JC-201804-citibike-tripdata.csv")
may_trip_df = pd.read_csv("Data/JC-201805-citibike-tripdata.csv")
jun_trip_df = pd.read_csv("Data/JC-201806-citibike-tripdata.csv")
jul_trip_df = pd.read_csv("Data/JC-201807-citibike-tripdata.csv")
aug_trip_df = pd.read_csv("Data/JC-201808-citibike-tripdata.csv")
sep_trip_df = pd.read_csv("Data/JC-201809-citibike-tripdata.csv")
oct_trip_df = pd.read_csv("Data/JC-201810-citibike-tripdata.csv")
nov_trip_df = pd.read_csv("Data/JC-201811-citibike-tripdata.csv")
dec_trip_df = pd.read_csv("Data/JC-201812-citibike-tripdata.csv")
```

```python
jan_trip_df.head()
```

```python
#create a list of df to use concat to combine data
frames2018 =[jan_trip_df,feb_trip_df,mar_trip_df,apr_trip_df,may_trip_df,jun_trip_df,jul_trip_df,aug_trip_df,sep_trip_df,
            oct_trip_df,nov_trip_df,dec_trip_df]
```

```python
#perform pd.concat and preview files
trip_data_2018 = pd.concat(frames2018)
```

```python
#will need to double check rows come out correctly
trip_data_2018
```

```python
#we are going to freate a column that changes the 0, 1 or 2 values to Unknown, male or female
def gender(row):
    if row["gender"] == 1:
        return 'Male'
    elif row["gender"] == 2:
        return 'Female'
    else:
        return "Unknown"
```

```python
#axis=1 specifier means that the application is done at a row rather than a column level
#save results into a new column
trip_data_2018["Gender"] = trip_data_2018.apply(lambda row: gender(row), axis=1)
```

```python
#check we have same number of rows in new column
trip_data_2018["Gender"].count()
```

```python
#drop original gender column now that we have a more descriptive Male/Female/Unknown Gender Column
trip_data_2018 = trip_data_2018.drop(columns=['gender'])
```

```python
#review our data
trip_data_2018.describe()
```

```python
#check for outliers cont for trips over 6 hours, less than 2 minutes, and birth year greater than 79
outlier_durations = trip_data_2018[(trip_data_2018.tripduration >= 21600) | (trip_data_2018.tripduration <= 120) | (trip_data_2018["birth year"] < 1940)]
outlier_durations.count()
```

```python
#calulate the percentage of outliers
outlier_durations.count()/trip_data_2018.tripduration.count()
```

```python
#we notice above that we have outliers in the min/max for tripduration, and outliers in the birth years
#set parameters for cleaning, make cutoff for trip duration to be 3 mins(120s) instead of 60 s
#set paramters for birth year to 1950? Don't believe many 69+ are riding bikes
trip_data_2018_clean = trip_data_2018[(trip_data_2018.tripduration <= 21600) & (trip_data_2018.tripduration >= 120) & (trip_data_2018["birth year"] > 1940)]
trip_data_2018_clean.describe(include='all')
```

```python
trip_data_2018_clean.head()
```

```python
#create bins for age group and the age group list for labels
bins = [1940, 1968,1978,1988,1998,2020]
age_groups = ["51+", "41-50", "31-40", "21-30", "<=20"]
#Add new column called Age Group to orginal purchase_data df
trip_data_2018_clean["Age Group"] = pd.cut(trip_data_2018_clean["birth year"], bins, labels=age_groups)
trip_data_2018_clean
```

```python
#send our file to a new csv file
trip_data_2018_clean.to_csv("Data/bike_trip_data_2018.csv")
```
